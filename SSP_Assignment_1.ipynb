```python
#Loading the texts from the file.
import os

forsik_dir = '/content'
text_files = []

for root, dirs, files in os.walk(forsik_dir):
    for file in files:
        if file.endswith('.txt'):
            text_files.append(os.path.join(root, file))

#for f in text_files:
    #print(f)

#len(text_files)
```

```python
#testing word localisation on one text
auto = open("/content/Automekaniker.txt", "r", encoding='ISO-8859-1').read()

atokens = nltk.word_tokenize(auto)
a=nltk.text.Text(atokens)
awords_tokenize = len(atokens)
# make lower case for searching
atokens2 = [s.lower() for s in atokens]


print(auto.count('mekaniker'))
auto.count('køretøj')
```

```python
#Using a for loop I used to count terms in HC Andersen, placing numbers in a dictionary.
words = ['forsikring', 'køretøj', 'redningshjælp', 'tyveri', 'udland']
counts_all = {w: 0 for w in words}

for p in words:
    counts_all[p] += atokens2.count(p)

print(counts_all)
```

```python
# Import required libraries
from nltk.stem.snowball import SnowballStemmer
from nltk.tokenize import word_tokenize

# Initialize stemmer
ss = SnowballStemmer("danish")
```

```python
#running through all the documents again using a modified version of the code I made for HC Andersen.
words = ['forsikring', 'køretøj', 'redningshjælp', 'tyveri', 'udland']
counts_all = {w: 0 for w in words}
all_docs = []

for l in text_files:
    print(l)
    counts_all = {w: 0 for w in words}
    with open(l, encoding='ISO-8859-1') as f:
        text = f.read().lower()
        tokens = nltk.word_tokenize(text)
        stemmed_words_ss = [ss.stem(word) for word in tokens]


        for p in words:
            counts_all[p] += stemmed_words_ss.count(p)
    print(counts_all)
    all_docs.append(counts_all)

print(all_docs)

```
