Defining emojis

```python
emojis = {"🤦🏻‍♂️": -2, "😁": 4, "😉": 3}
```

Loading the lexicon

```python
import pandas as pd

df = pd.read_csv('/content/danish-sentiment-lexicon/2_headword_headword_polarity.csv',
                header=None,
                usecols=[0, 4],
                names=['word', 'score'])

sentiment_dict = dict(zip(df['word'], df['score']))

sentiment_dict
```

```python
#loading nltk
import nltk
nltk.download('popular')
nltk.download('punkt_tab')
```

```python
#from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import word_tokenize
import re

def clean_text(text):
  text = text.lower()
  text = re.sub('[^a-zæøå\s]', '', text)
  tokens = word_tokenize(text)
  return tokens

clean_text(training[0])
```

```python
def analyze_sentiment(text):
    #Load tokens using cleaning function
    tokens = clean_text(text)

    #Count up scores for words and emojis
    scores = [sentiment_dict.get(w, 0) for w in tokens]
    emoji_score = [emojis.get(w, 0) for w in emojis if w in text]

    total_score = sum(scores) + sum(emoji_score)

    return total_score

#Running the function with the lexicon on the training text.
for post in training:
    total_score = analyze_sentiment(post)
    status = "Positive" if total_score > 0 else "Negative" if total_score < 0 else "Neutral"
    print(f"[{status}] Score: {total_score} | Post: {post}")
```

```python
for post in testing:
    total_score = analyze_sentiment(post)
    status = "Positive" if total_score > 0 else "Negative" if total_score < 0 else "Neutral"
    print(f"[{status}] Score: {total_score} | Post: {post}")
```

Kappa score

```python
import pandas as pd
from sklearn.metrics import cohen_kappa_score

file_path = "/content/Safira_Rachel_Lin_Nielsen_tweets_assignment.xlsx"
#Read sheets
train_df = pd.read_excel(file_path, sheet_name="Training")
test_df = pd.read_excel(file_path, sheet_name='Testing')
# Function to compute kappa
def evaluate(df, sheet_name):
    kappa_rule = cohen_kappa_score(
        df['label-human'],df['label-rule-based']
    )
    kappa_llm = cohen_kappa_score(
        df['label-human'],df['label-llm']
    )
    print(f'\n{sheet_name}')
    print(f'Human vs Rule-based Kappa: {kappa_rule:.3f}')
    print(f'Human vs LLM Kappa: {kappa_llm:.3f}')
# Evaluate both sheets
evaluate(train_df, 'Training Set')
evaluate(test_df, 'Testing Set')
```

```python
#Difference between human raters in testing
from sklearn.metrics import cohen_kappa_score

# Rater annotations
rater1 = ["Positive", "Positive", "Positive",
           "Negative", "Negative", "Negative",
           "Neutral", "Neutral", "Neutral"]

rater2 = test_df["label-human"]

# Calculate Kappa
kappa = cohen_kappa_score(rater1, rater2)

print(f"Kappa: {kappa:.4f}")
```
