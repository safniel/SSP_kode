```python
import sacrebleu



# Calculate corpus BLEU for model A
bleu_a = sacrebleu.corpus_bleu(
    candidate_a,
    [reference]
)

print(bleu_a)

# Calculate corpus BLEU for model B
bleu_b = sacrebleu.corpus_bleu(
    candidate_b,
    [reference]
)

print(bleu_b)
```

Calculating SBLEU

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

ref = [reference[0].split()]
can_a = candidate_a[0].split()
can_b = candidate_b[0].split()

# Without smoothing

bleu = sentence_bleu(ref, can_a)
print(f"BLEU model A: {bleu:.4f}")  # BLEU: 1.0000

bleu = sentence_bleu(ref, can_b)
print(f"BLEU model B: {bleu:.4f}")  # BLEU: 1.0000

# With smoothing (recommended)
bleu_smoothed = sentence_bleu(
    ref,
    can_a,
    smoothing_function=SmoothingFunction().method4
)
print(f"Smoothed BLEU model A: {bleu_smoothed:.4f}")

# With smoothing (recommended)
bleu_smoothed = sentence_bleu(
    ref,
    can_b,
    smoothing_function=SmoothingFunction().method4
)
print(f"Smoothed BLEU model B: {bleu_smoothed:.4f}")
```
